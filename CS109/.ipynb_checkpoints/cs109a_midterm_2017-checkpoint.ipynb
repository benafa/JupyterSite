### Delay vs time of departure 

```python
# define categories for departure- morning, afternoon/evening and night 
# morning: 4am-12pm
# afternoon: 12pm-8pm
# night: 8pm-4am 
import datetime
import time
four_am = time.strptime("04:00", "%H:%M")
twelve_pm = time.strptime("12:00", "%H:%M")
eight_pm = time.strptime("20:00", "%H:%M")

def find_period(x):
    x_time = time.strptime(x, "%H:%M:%S")
    if x_time >= four_am and x_time <= twelve_pm:
        return "MORNING"
    elif x_time >= twelve_pm and x_time <= eight_pm:
        return "AFTERNOON"
    else:
        return "NIGHT"
df['DEPARTURE_TIME_PERIOD'] = df['SCHED_DEP'].apply(lambda x: find_period(x))   
```

### Classification helper functions

```python
def classification_accuracy(df, classify):
    total = 0
    success = 0
    pred = []
    actual = []
    for index, row in df.iterrows():
        cat = row['DELAY_OR_NOT']
        row = row.drop("DELAY_OR_NOT")
        row = row.values.reshape(1,-1)
        pred_cat = classify(row)
        if pred_cat == cat:
            success = success + 1
        total = total+1
        pred.append(pred_cat)
        actual.append(cat)
    return success/total, pred, actual
```

```python
def get_confusion_matrix(df, classify):
    logistic_acccuracy = classification_accuracy(df, classify)
    #Confusion matric for test data on logistric model
    logistic_confusion_table = metrics.confusion_matrix(logistic_acccuracy[2], logistic_acccuracy[1])
    tn, fp, fn, tp = logistic_confusion_table.ravel()
    result_dict = {"Accuracies": ["True Negative", "False Positive", "False Negative", "True Positive"], "Count" : [tn, fp, fn, tp]}
    total_negative = tn + fp
    total_positive = tp + fn 
    tn, fp, fn, tp = tn/total_negative, fp/total_negative, fn/total_positive, tp/total_positive
    result_dict["Score"] = [tn, fp, fn, tp]
    result_df = pd.DataFrame.from_dict(result_dict).set_index("Accuracies")[['Score', "Count"]]
    return result_df
```

```python
#making ROC curves for this model from lab7
# altered to return fpr, tpr and thresholds 
from sklearn.metrics import roc_curve, auc

def make_roc(name, clf, ytest, xtest, ax=None, labe=5, proba=True, skip=0):
    initial=False
    if not ax:
        ax=plt.gca()
        initial=True
    if proba:#for stuff like logistic regression
        fpr, tpr, thresholds=roc_curve(ytest, clf.predict_proba(xtest)[:,1])
    else:#for stuff like SVM
        fpr, tpr, thresholds=roc_curve(ytest, clf.decision_function(xtest))
    roc_auc = auc(fpr, tpr)
    if skip:
        l=fpr.shape[0]
        ax.plot(fpr[0:l:skip], tpr[0:l:skip], '.-', alpha=0.3, label='ROC curve for %s (area = %0.2f)' % (name, roc_auc))
    else:
        ax.plot(fpr, tpr, '.-', alpha=0.3, label='ROC curve for %s (area = %0.2f)' % (name, roc_auc))
    label_kwargs = {}
    label_kwargs['bbox'] = dict(
        boxstyle='round,pad=0.3', alpha=0.2,
    )
    if labe!=None:
        for k in range(0, fpr.shape[0],labe):
            #from https://gist.github.com/podshumok/c1d1c9394335d86255b8
            threshold = str(np.round(thresholds[k], 2))
            ax.annotate(threshold, (fpr[k], tpr[k]), **label_kwargs)
    if initial:
        ax.plot([0, 1], [0, 1], 'k--')
        ax.set_xlim([0.0, 1.0])
        ax.set_ylim([0.0, 1.05])
        ax.set_xlabel('False Positive Rate')
        ax.set_ylabel('True Positive Rate')
        ax.set_title('ROC')
    ax.legend(loc="lower right")
    return ax, fpr, tpr, thresholds
```

### PCA

```python
def pca(n, X_train, y_train, X_validate, y_validate):    
    pca = PCA(n_components=n)

    pca.fit(X_train)   
    train_pca = pca.transform(X_train)
    test_pca  = pca.transform(X_validate)

    logitm = LogisticRegression(C = 100000)
    logitm.fit (train_pca, y_train)
    
    train_score = logitm.score(train_pca, y_train)
    test_score = logitm.score(test_pca, y_validate)
    
    return test_score

max_dim = 27

def cross_val_and_run_pca(num_folds, X_train, y_train, X_test, y_test, max_dim):
    
    interval = int(len(X_train) / num_folds) - 1
    pca_scores = []
    for n in range(1, max_dim):  # n-component PCA
        score = 0
        # begin cross-validation for given number of components
        for i in range(1, num_folds + 1):
            end_point = 0

            if i == num_folds:
                end_point = len(X_train)

            else:
                end_point = (i + 1) * interval

            X_train_validate_set = X_train[i * interval: end_point]
            y_train_validate_set = y_train[i * interval: end_point]
            
            X_train_train_set = np.concatenate((X_train[0:i*interval], X_train[end_point:]))
            y_train_train_set = np.concatenate((y_train[0:i*interval], y_train[end_point:]))
            score += pca(n, X_train_train_set, y_train_train_set, X_train_validate_set, y_train_validate_set)
        pca_scores.append(score / num_folds)
    return pca_scores

X_train = data_train.drop('DELAY_OR_NOT', 1).values
y_train = data_train["DELAY_OR_NOT"].values

X_test = data_test.drop('DELAY_OR_NOT', 1).values
y_test = data_test["DELAY_OR_NOT"].values

# use CV to find best number of components to use
pca_scores = cross_val_and_run_pca(5, X_train, y_train, X_test, y_test, max_dim)

best_performer = pca_scores.index(max(pca_scores)) + 1
print("best performing n-dimensional PCA: {}".format(best_performer))

print("Testing PCA performance (mean score): ", pca(best_performer, X_train, y_train, X_test, y_test))
```

```python
pca = PCA(n_components=best_performer)
pca.fit(X_train)   
train_pca = pca.transform(X_train)
logitm = LogisticRegression(C = 100000)
logitm.fit (train_pca, y_train)
# Accuracy of pca model
def classify_pca(val):
    expected = logitm.predict_proba(pca.transform(val))[0][1]
    if expected < 0.5:
        return 0.0
    else:
        return 1.0

pca_result = get_confusion_matrix(data_test, classify_pca)
pca_result
```

```python
# Use bootstrapping 
import random

X_train = data_train[best_logistic_predictors].values
X_test = data_test[best_logistic_predictors].values  

y_train = data_train["DELAY_OR_NOT"].values
y_test = data_test["DELAY_OR_NOT"].values


def bootstrap_sample(num_samples, max_index, dataset, y_set):
    indices = []

    for index in range(num_samples):
        indices.append(random.randint(0, max_index - 1))
    
    x = []
    y = []
    for index in indices:
        x.append(dataset[index])  
        y.append(y_set[index])
    return np.array(x), y

num_iterations = 100

beta_coeff = 0
beta_intercept = 0

beta_coeff_sd_list = []
beta_intercept_sd_list = []

for n in range(num_iterations):
    
    X_train_sample, y_train_sample = bootstrap_sample( len(X_train), len(X_train), X_train, y_train)
    y_train_sample = np.array(y_train_sample).reshape((-1, ))
    
    logitm = LogisticRegressionCV(fit_intercept=True, Cs=lambdas, penalty='l2', multi_class = 'ovr')
    logitm.fit (X_train_sample, y_train_sample)
    
    # The coefficients
    beta_coeff += logitm.coef_
    beta_intercept += logitm.intercept_
    

    beta_coeff_sd_list.append(logitm.coef_[0,:].tolist())
    beta_intercept_sd_list.append(logitm.intercept_.tolist())

beta_coeff_sd_list = np.array(beta_coeff_sd_list)

beta_intercept_sd_list = np.array(beta_intercept_sd_list)
print("Done.")

beta_coeff /= num_iterations
beta_intercept /= num_iterations

beta_coeff_sd = np.std(beta_coeff_sd_list, axis = 0)
beta_intercept_sd = np.std(beta_intercept_sd_list, axis = 0)

print(list(zip(beta_coeff_sd, best_predictor_set_b[0])))
print(beta_intercept_sd)
```

```python
from scipy.stats import t

t_interval = t.interval(0.95, len(y_train) - 1)
print(t_interval)

significant = lambda x: x < t_interval[0] or x > t_interval[1]

signif_count = 0

signif_intercept = 0

betas = beta_coeff[0,:].tolist()

index = []
for n in range(len(betas)):
    beta_t = betas[n] / beta_coeff_sd[n]
    if significant(beta_t):
        signif_count += 1
        index.append(n)

        
signif_intercept += significant(beta_intercept[0] / beta_intercept_sd[0])
print(signif_count)
print(signif_intercept)

for i in index:
    print(best_logistic_predictors[i])

```

```python
# Backward selection 
from statsmodels.tools import add_constant
from statsmodels.api import OLS

X_train = data_train.drop('ARRIVAL_DELAY', axis=1)
y_train = data_train['ARRIVAL_DELAY']
X_test = data_test.drop('ARRIVAL_DELAY', axis=1)
y_test = data_test['ARRIVAL_DELAY']

def get_score(X_train, y_train, X_test, y_test):
    regression_model = LinearRegression(fit_intercept=True)
    model = regression_model.fit(X_train, y_train)
    return model.score(X_test, y_test)

all_predictors = list(data_train.drop('ARRIVAL_DELAY', 1).columns)
X_train = data_train[all_predictors].values
predictors = [(all_predictors, get_score(X_train, y_train, X_train, y_train))]

for k in range(len(all_predictors), 1, -1):
    best_k_predictors = predictors[-1][0]
    scores = []
    for predictor in best_k_predictors:
        k_minus_1 = list(set(best_k_predictors) - set([predictor]))
        X_train = data_train[k_minus_1].values
        X_test = data_test[k_minus_1].values

        scores.append(get_score(X_train, y_train, X_test, y_test))
    
    best_k_minus_1 = list(set(best_k_predictors) - set([best_k_predictors[np.argmax(scores)]]))
    predictors.append((best_k_minus_1, np.max(scores)))
best_predictor_set_b = sorted(predictors, key=lambda t: t[1])[61]

X_train = data_train[best_predictor_set_b[0]].values
X_test = data_test[best_predictor_set_b[0]].values  

regression_model = LinearRegression(fit_intercept=True)
model = regression_model.fit(X_train, y_train)
print(model.score(X_test, y_test))
print('best predictor set: {}\nscores: {}\ntest R^2: {}'.format(best_predictor_set_b[0], best_predictor_set_b[1], regression_model.score(X_test, y_test)))
```

### Time Period

```python
# define categories for departure- morning, afternoon/evening and night 
# morning: 4am-12pm
# afternoon: 12pm-8pm
# night: 8pm-4am 
import datetime
import time
four_am = time.strptime("04:00", "%H:%M")
twelve_pm = time.strptime("12:00", "%H:%M")
eight_pm = time.strptime("20:00", "%H:%M")

def find_period(x):
    x_time = time.strptime(x, "%H:%M:%S")
    if x_time >= four_am and x_time <= twelve_pm:
        return "MORNING"
    elif x_time >= twelve_pm and x_time <= eight_pm:
        return "AFTERNOON"
    else:
        return "NIGHT"
new_df['DEPARTURE_TIME_PERIOD'] = df['SCHED_DEP'].apply(lambda x: find_period(x))   
```
